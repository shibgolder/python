```python
# https://www.youtube.com/watch?v=Z81JW1NTsO8&list=PLQVvvaa0QuDfefDfXb9Yf0la1fPDKluPF&t=12
#  Drawing simple lines...
import matplotlib.pyplot as plt

x = [1,2,3]
y = [4,7,6]

x2 = [1,2,3]
y2 = [11,14,16]

plt.plot(x,y, label="1st line")
plt.plot(x2,y2, label="2nd line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("MPlot Line Graph")
plt.legend()
plt.show()
```

```python
#  Drawing Bar charts...
import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [4,7,6,5,6]

x2 = [1,3,5,7,9]
y2 = [8,5,6,7,2]

plt.bar(x,y, label="Bar1", color='r')
plt.bar(x2,y2, label="bar2", color='b')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("MPlot Bar Graph")
plt.legend()
plt.show()
```

```python
#  Drawing scatter chart...
import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [4,7,6,5,6]

plt.scatter(x,y, label="scatter", color='r', marker='o', s=100)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("MPlot Scatter Graph")
plt.legend()
plt.show()
```
